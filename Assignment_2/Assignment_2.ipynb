```python
# we will write a function to move columns 
def moving_cols(df1, cols_to_move, new_loc):
    col = df1.pop(cols_to_move)  # Remove column from the DataFrame
    df1.insert(new_loc, cols_to_move, col)  # Insert removed column at a new position
```

```python
plt.figure(figsize=(4,3))
sns.boxplot(x=df['price'])
plt.title("Boxplot of Revenue")
plt.show()
```

```python
Q1 = df1['price'].quantile(0.25)
Q3 = df1['price'].quantile(0.75)
IQR = Q3 - Q1

lowerp = Q1 - 1.5 * IQR
upperp = Q3 + 1.5 * IQR

outliers = df1[(df1['price'] < lower) | (df1['price'] > upper)]
outliers.head(15)
```

```python
Q1 = df1['area'].quantile(0.25)
Q3 = df1['area'].quantile(0.75)
IQR = Q3 - Q1

lowera = Q1 - 1.5 * IQR
uppera = Q3 + 1.5 * IQR

outliers = df1[(df1['area'] < lower) | (df1['area'] > upper)]
outliers.head(15)
```

```python
df_no_outliers = df1[(df1['price'] >= lowerp) & (df1['price'] <= upperp) & (df1['area'] >= lowera) & (df1['area'] <= uppera)]
print("Original shape: ",df1.shape)
print("After removing outliers: ",df_no_outliers.shape)
```

```python
df_no_outliers[['area']].head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df_scaled = df_no_outliers[['area']].copy()

df_scaled[['area']] = scaler.fit_transform(df_scaled)

df_scaled.head()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(df_no_outliers[['price','area']].corr(), 
            annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap (Before PCA)")
plt.show()
```

```python
from sklearn.decomposition import PCA

X = df_no_outliers[['price','area']]

pca = PCA(n_components=2)
principal_components = pca.fit_transform(X)

print("Explained Variance Ratio:", pca.explained_variance_ratio_)
```

```python
plt.figure(figsize=(6,4))
plt.scatter(principal_components[:,0], principal_components[:,1])
plt.title("PCA Projection")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()
```
